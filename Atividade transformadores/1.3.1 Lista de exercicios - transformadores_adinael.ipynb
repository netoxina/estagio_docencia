## Questão 4

Um transformador de 10 KVA, 4800/240 V é ensaiado a vazio e a curto-circuito:

Ensaio à  Vazio (Lado de BT): 60 W; 240 V; 1,5 A

Ensaio em Curto-Circuito (lado de AT): 180 W; 180 V; corrente nominal.

a) Determine o circuito equivalente referido para o primário e secundário

b) Determine a regulação de tensão do transformador à plena carga e fator de potência unitário.

```python
# Solução Q4 
# Ensaio a vasio
a = 4800/240
v1 = 240 # V
p1 = 60 # W
i1 = 1.5 # A

# Ensaio a curto circuito
v2 = 180 # V
i2 = 10000/4800 # A
p2 = 180 # 

# a)
# Para o ensaio a vazio temos:
fp1 = p1/(v1*i1)
teta1 = m.acos(fp1*m.pi/180)
ye = i1/v1
rc = 1/(ye*m.cos(-teta1))
xm = -1/(ye*m.sin(-teta1))

# Para o ensaio a curto circuito temos:
fp2 = p2/(v2*i2)
teta2 = m.acos(fp2)
z = v2/i2
req = z*m.cos(teta2)
xeq = z*m.sin(teta2)
```

```python
# Referenciado ao primario
# Para representar o circuito equivalente no prinario e nescessario converter os valores do ramo de excitação para o mesmo.
rcp = rc*a**2
xmp = xm*a**2
print("Rc.p = {:.2f}; Xm.p = {:.2f}".format(rcp,xmp))
print("Req = {:.2f}; Xeq = {:.2f}".format(req,xeq))
```

```python
# Referenciado ao secundario
# Para obter o circuito equivalente no secundario basta converter a resistencia e a reatância do primario para o secundario.
reqs = req/a**2
xeqs = xeq/a**2
print("Req.s = {:.2f} Xeq.s = {:.2f}".format(reqs,xeqs))
print("Rc = {:.2f}; Xm = {:.2f}".format(rc,xm))
```

```python
# b) Regulagem de tensão a plena carga
# Calculo do Vp/a com FP = 1 
i = 10000/240
fi = m.acos (1)
vpa = 240 + complex(reqs*i*m.cos(fi),reqs*i*m.sin(fi)) + complex(xeqs*i*m.cos(fi+m.pi/2),xeqs*i*m.sin(fi+m.pi/2))
rt = ((abs(vpa) - 240)/240)*100
print("RT = {:.2f}%".format(rt))
```

## Questão 5

Um transformador de 100 KVA, 60 Hz, 12000/240 V é ensaiado a vazio e a curto-circuito:

Ensaio à Vazio: 480 W; 240 V; 8,75 A 

Ensaio de Curto-Circuito: 1200 W, 600 V, Corrente nominal

Determine a regulação de tensão para carga nominal com Fator de Potência 0.8 em atraso.

```python
# Solução Q5
a = 1200/240
pcc = 1200
vcc = 600
icc = 100000/1200

# Calculo do Req e da Xeq a partir do ensaio em curto circuito.
z = vcc/icc
fi = m.acos(0.8)
req = z*m.cos(fi)
xeq = z*m.sin(fi)
# Convertendo para o lado de baixa temos:
reqs = req/a**2
xeqs = xeq/a**2
i = 100000/240
vpa = 240 + complex(reqs*i*m.cos(fi),reqs*i*m.sin(fi)) + complex(xeqs*i*m.cos(fi+m.pi/2),xeqs*i*m.sin(fi+m.pi/2))
rt = ((abs(vpa) - 240)/240)*100
print("RT = {:.2f}%".format(rt))
```

## Questão 7

Um transformador monofásico 250 KVA, 24200/220 V fora ensaiado apresentando os seguintes dados:

Ensaio a vazio (BT): 1200W, Tensão nominal, 16 A

Ensaio de curto-circuito (AT): 1700 W, 600 V, Corrente nominal

Determine o circuito equivalente e a regulação de tensão do transformador quando ele opera com uma carga 85% da nominal e fator de potência de 0.92 indutivo.

```python
# Solução Q7
fi = m.acos(0.92)
a =24200/220
# Dados a vazio
pvz = 1200 # W
vvz = 220 # V
ivz = 16 # A
# Dados a curto circuito
pcc = 1700 # W
vcc = 600 # V
icc = 250000/24200 # A
# Do ensaio a vazio temos:
ye = ivz/vvz
rc = 1/(ye*m.cos(-fi))
xm = -1/(ye*m.sin(-fi))
# Do ensaio a curto circuito temos:
z = vcc/icc  
req = z*m.cos(fi)
xeq = z*m.sin(fi)
# Regulação de tensão
reqs = req/a**2
xeqs = xeq/a**2
v1 = 220*0.85
i = 250000/v1
vpa = (220*0.85) + complex(reqs*i*m.cos(fi),reqs*i*m.sin(fi)) + complex(xeqs*i*m.cos(fi+m.pi/2),xeqs*i*m.sin(fi+m.pi/2))
rt = ((abs(vpa) - v1)/v1)*100
print("Req.s = {:.4f} Xeq.s = {:.4f}".format(reqs,xeqs))
print("Rc = {:.2f}; Xm = {:.2f}".format(rc,xm))
print("RT = {:.2f}%".format(rt))
```
