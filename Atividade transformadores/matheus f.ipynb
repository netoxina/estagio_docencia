```octave
% 8° - questão
%Letra - A

a = 8000/277;
Rs = 0.005; %ohm
Xs = 0.006; %ohm
Xm = 10*1000; %Obs: Valor de resistência muito grande, logo, a corrente tende ir pelo caminho mais fácil e não passará por aqui.
Rp = 5; %ohm
Xp = 6; %ohm
Rc = 50*1000; %Obs: Valor de resistência muito grande, logo, a corrente tende ir pelo caminho mais fácil e não passará por aqui.
Reqs = (Rp/a^2) + Rs; %ohm - Resistência equivalente para o lado de baixa 
Xeqs = (Xp/a^2) + Xs; %ohm - Reatância equivalente para o lado de baixa 

%Letra - B

VAbases = 100*1000; %VA
Vbases = 277; %V
Rbase = (Vbases^2)/VAbases;
Zbts = (Rs + j*Xs)/Rbase; %pu - Circuito equivalente para o lado de baixa em unidades por transformador 
Zbtp = (a^2)*Zbts; %pu - Circuito equivalente para o lado de alta em unidades por transformador 
disp (['Resistência equivalente para o lado de baixa  = ' num2str(Reqs)]);
disp (['Reatância equivalente para o lado de baixa = ' num2str(Xeqs)]);
disp (['Circuito equivalente para o lado de baixa em unidades por transformador  = ' num2str(Zbts)]);
disp (['Circuito equivalente para o lado de alta em unidades por transformador  = ' num2str(Zbtp)]);

%Letra - C

S = 100*1000;
Vn = 277; %V
ft = -0.85; %atrasado
In = S/Vn; %Tensão nominal
Isr = In*cos(ft);
Isi = In*sin(ft)*j;
Vpa = Vn + (Reqs + j*Xeqs)*(Isr + Isi);
Rt = ((Vpa - Vn)/Vn)*100;
Rt1 = abs(Rt); % "Rt1%"  %Regulação de tensão em porcentagem
disp (['Tensão de entrada = ' num2str(Vn)]);
disp (['Regulação de tensão em porcentagem = ' num2str(Rt1)]);

%Letra - D

Pc = Reqs*In**2; %Perdas no cobre
Pn = (Vpa**2)/Rc; %Perdas no núcleo
disp (['Perdas no cobre = ' num2str(Pc)]);
disp (['Perdas no núcleo = ' num2str(Pn)]);

%Letra - E

Pe = Vn*In*ft; %Perdas de entrada

Ps = Pe + Pc + Pn; %Perdas de saída
N = abs((Ps/Pe)*100); %Eficiência 
disp (['Eficiência = ' num2str(N)]);
```
