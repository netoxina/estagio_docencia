## Questão 1

Uma fonte de impedância interna de 512 ohms alimenta um alto-falante com uma impedância de apenas 8 ohms. Nestas condições é de se esperar que a potência fornecida ao alto-falante seja muito pequena, diante a divisão de tensão entre as cargas. Desta maneira, calcule:



a) Valor da potência fornecida a carga de acordo com a figura (a)

b) Na figura (b) é possível observar a introdução de um transformador para casamento de impedâncias. Calcule a nova potência fornecida ao alto-falante.

c) Compare as potências antes e depois de se introduzir o transformador

```octave
%% Solução Q1

% A
Rp = 512; %Impedância no primário
Rs = 8; %Impedância no secundário
Rt = Rp + Rs; %Impedância total
Vg = 120; %Tensão de entrada
If = Vg/Rt; %Corrente da fonte
P = (I^2)*Rs %Potência fornecida ao alto-falante

% B
Np = 8; %Número de espiras no primário
Ns = 1; %Número de espiras no secundário
a = Np/Ns; %Relação entre o número de espiras
Zc = 8; %Impedância no secundário
Zp = (a^2)*Zc; %Impedância no primário
Zt = Rp + Zp; %Impedância total
Ip = Vg/Zt; %Corrente da fonte
Pb = (Ip^2)*Zp %Nova potência fornecida

% C
%Após introduzir o transformador, nota-se que a potência aumenta, devido ao aumento da impedância.
```

## Questão 8

Um transformador de distribuição de 100 kVA, 8000/277 V tem as seguintes resistências e reatâncias:




a) Encontre o circuito equivalente do equipamento referente ao lado de baixa.

b) Encontre o circuito equivalente por unidade desse transformador

c) Assuma que o transformador alimente uma carga nominal em 277 V e FP 0,85 atrasado. Qual é sua tensão de entrada? Qual é sua regulação de tensão?

d) Quais são as perdas no cobre e no núcleo desse transformador nas condições do item (c)

e) Determine a eficiência do transformador nas mesmas condições

```octave
% Solução Q8

Rp = 5; %ohm
Rs = 0.005; %ohm
Xp = 6; %ohm
Rc = 50; %kiloohm
Xs = 0.006; %ohm
Xm = 10000; %ohm

% A
a = 8000/277; %Razão de transformação
%Realizando os cálculos para o lado de baixa tensão:
Rp1 = Rp/(a^2); %ohm
Xp1 = Xp/(a^2); %ohm
Rc1 = Rc/(a^2); %ohm
Xm1 = Xm/(a^2); %ohm

% B
Is = 100000/277; %Corrente no lado secundário
Zb = 277/Is; %Impedância base do lado primário
%Logo, o circuito ficará assim:
Rp2 = Rp1/Zb; %ohm
Xp2 = Xp1/Zb; %ohm
Rc2 = Rc1/Zb; %ohm
Xm2 = Xm1/Zb; %ohm

% C
Req = 0.011; %Resistência equivalente
Xeq = 0.0132; %Reatância equivalente
Vs = 277; %Tensão secundária
Vp1 = Vs + (Req + Xeq)*Is %Tensão primária referida ao lado secundário num ângulo de 31,8º
VR = ((Vp1-Vs)/Vs)*100 %Regulação de tensão em %

% D
Ps = 100000*cos(31.8); %Perda na saída em kW
Pc = (Is^2)*Req %Perda no cobre em W
Pn = (Vp1^2)/Rc %Perda no núcleo em W

% E
eficiencia = ((Ps)/(Ps+Pc+Pn))*100 %Eficiência do transformador em %
```

## Questão 9

O enrolamento secundário de um transformador real tem uma tensão de terminal de $v_{s}(t) = 282,8 sen (377t)$ V. A relação de espiras do transformador é 100:200 ($a = 0,50$). Se a corrente do secundário no transformador for $i_{s}(t) = 7,07 sen (377t = 36,87°)$ A, qual será a corrente do primário desse transformador? Quais são sua regulação de tensão e sua eficiência? As impedâncias do transformador, referidas ao lado do primário, são





```octave
%Solução Q10

Rc = 300; %ohm
Req = 0.2; %ohm
Xm = 100; %ohm
Xeq = 0.8; %ohm
a = 0.5;

Vs = 282.8/sqrt(2); %Tensão no secundário
Is = 7.07/sqrt(2); %Corrente no secundário num ângulo de 36,87º
Vs1 = Vs*a; %Tensão no secundário se referindo ao lado primário
Is1 = Is/a; %Corrente no secundário se referindo ao lado primário, num ângulo de 36,87º
Vp = Vs1 + Is1*(Req+Xeq); %Tensão no lado primário num ângulo de 2,8º
Ic = Vp/Rc; %A
Im = Vp/Xm; %A
Ie = Ic+Im; %Corrente de excitação num ângulo de -68,8º, A
Ip = Is1 + Ie %Corrente primária, num ângulo de 40º
VR = ((Vp-(a*Vs))/(a*Vs))*100 %Regulação de tensão em %
Pe = Vp*Ip*0.734; %Potência de entrada, ângulo de 42,8º
Ps = Vs*Is*0.8; %Potência de saída, ângulo de 36,87º
eficiencia = (Ps/Pe)*100 %Eficiência do transformador em %
```
