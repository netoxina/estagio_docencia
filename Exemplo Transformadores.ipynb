## Exemplo 1
Para o transformador de núcleo de ferro da figura abaixo, determine:

a) Módulo da corrente no primário e da tensão aplicada no secundário

b) Resistência de entrada do transformador

```octave
R = 2000 % Ohm
Np = 40 % voltas
Ns = 5 % voltas
Is = 0.1 % A

```

Como primeiro passo para encontrar a corrente, tensão , assim como a impedância referente ao primário, devemos encontrar a relação de espiras, convencionalmente calculada como   $ a = \dfrac{Na}{Nb}$ ,

sendo Na o número de espiras do lado de alta e Nb o de baixa.

Assim, para calcularmos tensão e corrente no lado de alta basta:

$V_{p}= a * V_{s}$ ,

$I_{p}= \dfrac{1}{a} * I_{s}$,

```octave
a = Np/Ns % adimensional
Ip = Is/a % A
Vs= Is* R % V
Vp= Vs*a % V
```

b) Possuindo a relação de espiras (a) é possível ainda determinar os valores de impedância vista do primário. Para isso é necessário rebater a impedância do lado de baixa para o de alta, fazemos:

```octave
Rp = (a^2)*R % Ohm
```

Note que para calcular a resistência do lado alta, lado de maior impedância, devemos

$Z_{alta} = a^{2} * Z_{baixa}$

$Z_{baixa} = (\dfrac{1}{a^{2}}) * Z_{baixa}$

Neste caso a impedância do lado de alta está no primário, por isso $Z_{alta} = R_{p}$ 

## Exemplo 2

Determinar para o transformador da figura abaixo:

a) $R_{eq}$ e $X_{eq}$ circuito equivalente

b) Determinar os módulos de tensões $V_{c}$ e $V_{g}$

c) Determinar o módulo de tensão de saída $V_{c}$ caso as perdas no transformador não existissem. Quais observações podem ser feitas ao comparar com o resultado de b) ?

o primeiro passo para encontrarmos o circuito equivalente do transformador é rebatermos as impedâncias para um dos lados. Dessa maneira, encontraremos o $a$ e em seguida realizaremos o cálculo do ponto de vista do primário (lado de Alta).

```octave
Rp = 1 % Ohm
Rs = 1 % Ohm
Xp = 2 % Ohm
Xs = 2 % Ohm
Rc = 60 % Ohm
Ip = 10 % A
a=2/1 % adimensional
```

```octave
Req = Rp + (a^2)*Rs
Xeq = Xp + (a^2)*Xs
```

b) Para determinar a tensão $V_{c}$ será necessário encontrar a corrente do secundário, como já temos $a$, basta

```octave
Is = a*10 % A
Vc= Is*Rc % V
```

Para $V_{g}$ utilizaremos o circuito equivalente e a 2ª lei de kirchhoff (lei das malhas)

```octave
Rcp=(a^2)*Rc
Vg= (Req+Rcp+j*Xeq)*Ip
```

Se ainda quisermos encontrar o ângulo de fase

```octave
VG=abs(Vg) % módulo de Vg em V
fi=angle(Vg) % radianos
fi2=rad2deg(fi) % graus
```

c) Nessa etapa iremos desprezar todas as perdas do transformador, para modelar isso devemos desprezar $R_{eq}$ e $X_{eq}$, dessa forma o novo valor de Vg é calculado

```octave
Vg= Ip*Rcp % V
```

Ao comparar com o resultado anterior, devido as perdas nossa fonte deve fornecer uma tensão superior a 2400 V para que esta tenha sua demanda atendida. Além disso, perde-se toda a parte reativa da tensão devido a carga ser puramente resistiva.
